```python
%%px
from astroML.datasets import fetch_sdss_specgals
import numpy as np
data=fetch_sdss_specgals()

#put magnitudes in a matrix
X=np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y= data['z']

#down-sample the data for speed
X=X[::10]
y=y[::10]
```

```python
#%px from sklearn.ensemble import RandomForestRegressor
#%px from sklearn.cross_validation import cross_val_score, StratifiedKFold

dview.scatter('a',range(5,25))

dview.execute('''
score=[]
for num in a:
    # define the model with the given value of C
    model = RandomForestRegressor(n_estimators=num, max_depth=3)
    #compute the scores via cross-validation
    scores = cross_val_score(model, X, y, scoring= 'r2',cv=3)
    
    #print the mean of the cross-validation scores
    score += [np.mean(scores)]''')

score=dview.gather('score')
print(score)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(5,25),score)
plt.xlabel('Number of Estimators')
plt.ylabel('Mean score')
plt.title('Change in mean score when # of estimators is varied')
```
